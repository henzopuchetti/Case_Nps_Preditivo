## **Importação e Scripts**

##### Importação das bibliotecas para utilização na analise. Juntamente com a importação da base de dados e visualização do head da mesma.

```python
import pandas as pd # Manipulação e analise dos dados
import numpy as np # Manipulação de operações matematicas.
import matplotlib.pyplot as plt # Criação de graficos.
import seaborn as sns # Manipulação e visualização de estatisticas mais avançadas e estilizadas.


df = pd.read_csv('desafio_nps_fase.csv') # Importação da base de dados
df.head() # Leitura do cabeçalho
```

##### Explorando mais os dados para conseguir entender mais sobre a base de dados, com a quantidade de dados e os tipos das variaveis presentes.

```python
df.info()
df.describe()
```

##### Divisão da nota do NPS em grupos (Detrator, Neutro e Promotor) para conseguir ver a satisfação dos clientes.

```python
# Divisão de grupos e definição de score para que consigamos classifica-los e analisa-los de forma mais precisa
def classificar_nps(score):
    if score <= 6:
        return 'Detrator'
    elif score <= 8:
        return 'Neutro'
    else:
        return 'Promotor'

df['nps_group'] = df['nps_score'].apply(classificar_nps)
```

```python
df['nps_group'].value_counts(normalize=True)
```

##### De acordo com esses dados é possivel analisar que em sua maioria os clientes são detratores, assim percebemos que há problemas com a experiência do cliente.

##### Analisando e comparando por grupo, para entender os padrões que podem estar ocorrendo para que os clientes virem detratores

```python
df.groupby('nps_group')[[
    'delivery_delay_days',
    'complaints_count',
    'customer_service_contacts',
    'resolution_time_days'
]].mean()
```

##### Podemos identificar que de acordo com a analise desses dados, os detratores tem um comportamento padrão com relação ao atraso na entrega, reclamações e com isso um aumento ao contatar o suporte. Logo gerando uma insatisfação no serviço prestado.


##### Analisando para ver se achamos o ponto de ruptura do NPS

```python
df.groupby('complaints_count')['nps_score'].mean()
```

##### Com base no ponto de ruptura, podemos identificar que a partir do 2 dia de atraso, o NPS cai drasticamente.
